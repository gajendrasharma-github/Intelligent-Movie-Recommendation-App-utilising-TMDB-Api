```python
model_data.dropna(inplace=True)
```

### Cleaning the Genres Column

```python
model_data.iloc[0, 1]
```

```python
import ast
def convert(obj):
    obj=ast.literal_eval(obj)
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
```

```python
convert(model_data.iloc[0, 1])
```

```python
model_data['genres']=model_data['genres'].apply(convert)
```

### Cleaning the Keywords Columns

```python
model_data['keywords']=model_data['keywords'].apply(convert)
```

```python
model_data.head()
```

### Cleaning Cast Column

```python
def convert(obj):
    obj=ast.literal_eval(obj)
    counter=0
    L=[]
    for i in obj:
        if counter >=3:
            break
        else:
            L.append(i['name'])
            counter=counter+1
    return L
```

```python
convert(model_data.cast[0])
```

```python
model_data['cast']=model_data['cast'].apply(convert)
```

```python
model_data.head(2)
```

### Cleaning Crew

```python
ast.literal_eval(model_data['crew'][0])
```

```python
def convert(obj):
    obj=ast.literal_eval(obj)
    name=[]
    for i in obj:
        if i['job']=='Director':
            name.append(i['name'])
    return name
```

```python
convert(model_data['crew'][0])
```

```python
model_data['crew']=model_data['crew'].apply(convert)
```

```python
model_data.head()
```

### Cleaning the overview column or rather transforming the overview column

```python
model_data['overview']=model_data['overview'].apply(lambda x: x.split())
```

### Transforming the Genres, Cast, Crew and Keywords column to remove space

```python
model_data['genres']= model_data['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
model_data['keywords']= model_data['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
model_data['cast']= model_data['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
model_data['crew']= model_data['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
model_data.head()
```

```python
model_data['tags']= model_data['genres'] + model_data['keywords']+ model_data['overview'] + model_data['cast'] + model_data['crew']
```

```python
model_data.head()
```

```python
preprocessed_data=model_data[['movie_id', 'title', 'tags']]
preprocessed_data.head()
```

```python
preprocessed_data['tags']=preprocessed_data['tags'].apply(lambda x: ' '.join(x))
```

```python
preprocessed_data.head()
```

```python
preprocessed_data['tags'][0]
```

```python
preprocessed_data['tags']=preprocessed_data['tags'].apply(lambda x: x.lower())
```

```python
preprocessed_data.head()
```

### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(preprocessed_data['tags']).toarray()
```

```python
vectors[0].tolist()
```

```python
cv.get_feature_names_out().tolist()
```

```python
len(cv.get_feature_names_out())
```

### Applying Stemming to get root words and solve the repeating similar words issue

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
preprocessed_data['tags'] = preprocessed_data['tags'].apply(stem)
```

```python
preprocessed_data['tags'][0]
```

```python
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(preprocessed_data['tags']).toarray()
```

```python
cv.get_feature_names_out().tolist()
```

```python
vectors.shape
```

#### Now we will calculate the Cosine Distance

```python
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity
```

```python
pickle.dump(similarity, open('similarity.pkl', 'wb'))
```

```python
def recommend(movie):
    movie_index = preprocessed_data[preprocessed_data['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(preprocessed_data.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```
